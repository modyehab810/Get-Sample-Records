### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Read The Data

```python
df = pd.read_excel("sales_data.xlsx")
```

```python
df.shape
```

```python
df.head()
```

### Replace Spaces From Columns With _ 

```python
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

### Set Index 

```python
df.set_index("Row_ID", inplace=True)
```

```python
df.head()
```

### Get Random Sample

```python
# Get 37% sample Data From Original Data
random_sample = df.sample(frac=0.37, random_state=3)
random_sample.shape
```

```python
random_sample
```

```python
# on The Fly: Sort The random_sample According to The Index
random_sample.sort_index()
```
